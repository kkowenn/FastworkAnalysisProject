# CountSubCat.xlsx

```python
import pandas as pd

user = pd.read_excel('aboutUser(groupbySAMEName+Cat+subCat).xlsx')
```

```python
user.head()
```

```python
grouped = user.sort_values(by='Total_user_gain', ascending=False)
summary = grouped.groupby('namefreelance').agg({'Total_user_gain': 'sum'})

# Saving to an Excel file
file_path = 'edited/user_sorted_grouped.xlsx'
summary.to_excel(file_path, index=True)
```

```python
# Counting sub-categories for each category
category_counts = grouped.groupby(['namefreelance', 'category_id']).size().unstack(fill_value=0)

# Adding the category counts to the summary DataFrame
summary = summary.join(category_counts)

# Calculating the total number of sub-categories for each freelancer
summary['Total_Sub_Categories'] = category_counts.sum(axis=1)

# Saving to an Excel file
file_path = 'edited/user_sorted_grouped.xlsx'
summary.to_excel(file_path, index=True)
```

# LoyalVSNew

To analyze the topic "New customer vs Loyal customer" and determine whether it's more effective to focus on gaining customer loyalty or marketing to attract new customers, you should consider the following columns from your dataset:

### Posts Data:
1. **Post_ID** - Unique identifier for each post.
2. **category_id** - Category of the freelance task.
3. **sub_category** - Sub-category of the freelance task.
4. **title** - Title of the freelance task.
5. **namefreelance** - Name of the freelancer.
6. **rating_comment** - Rating of the freelancer.
7. **Num_comment** - Number of comments.
8. **percentsuccess** - Success rate of the freelancer.
9. **cost** - Cost of the task.
10. **num_sold** - Number of tasks sold.
11. **repeat hired** - Number of times the freelancer was hired again.

### User Data:
1. **namefreelance** - Name of the freelancer.
2. **category_id** - Category of the freelance task.
3. **sub_category** - Sub-category of the freelance task.
4. **cost_per_task** - Cost per task.
5. **Total_Sum_User_Sold** - Total number of tasks sold by the user.
6. **Total_Hire_Repeat_User** - Total number of repeat hires by the user.
7. **Total_user_gain** - Total earnings of the user.
8. **RateGetting_loyaltCustomer** - Rate of gaining loyal customers.
9. **Total_user_gain_from_loyaltyCustomer** - Total earnings from loyal customers.
10. **User_Rating_Comment** - User rating.
11. **User_Num_Comment** - Number of user comments.
12. **RateOfGetingReview** - Rate of getting reviews.

### Analysis Focus:
- **Customer Loyalty Analysis:**
  - **repeat hired**: Number of times a freelancer was hired again.
  - **RateGetting_loyaltCustomer**: Rate of gaining loyal customers.
  - **Total_Hire_Repeat_User**: Total number of repeat hires by the user.
  - **Total_user_gain_from_loyaltyCustomer**: Total earnings from loyal customers.
  - **User_Rating_Comment**: User rating as a potential indicator of customer satisfaction and loyalty.
  - **RateOfGetingReview**: Rate of getting reviews, which might correlate with customer satisfaction and loyalty.

- **New Customer Acquisition Analysis:**
  - **num_sold**: Number of tasks sold.
  - **Total_Sum_User_Sold**: Total number of tasks sold by the user.
  - **cost_per_task**: Cost per task to understand the price attractiveness.
  - **PromotionalWordCount**: Number of promotional words in the title to gauge marketing effectiveness.
  - **NumFamousWord**: Number of famous words in the title to gauge visibility and attraction.
  - **lengthTitle**: Length of the title to analyze its impact on customer acquisition.

### Key Metrics:
- **Average Cost per Task**: To determine the cost-effectiveness of tasks.
- **Total_Num_Sold**: Total number of tasks sold to new customers.
- **Total_Hire_Repeat**: Total number of repeat hires.
- **AVG_Num_Sold**: Average number of tasks sold.
- **AVG_Hire_Repeat**: Average number of repeat hires.
- **AVG_Hire_Repeat_Percentage**: Percentage of repeat hires.
- **Total_ZeroNum_SoldAndRepeatedHire**: Total tasks with zero sales and repeat hires.
- **Total_ZeroNum_SoldAndRepeatedHire_Percentage**: Percentage of tasks with zero sales and repeat hires.
- **Max_Num_Comment**: Maximum number of comments.
- **AVG_Num_Comment**: Average number of comments.

### Steps for Analysis:
1. **Segment Customers**: Divide customers into new and loyal categories based on repeat hires.
2. **Analyze Costs**: Compare the average cost per task for new vs. loyal customers.
3. **Evaluate Performance**: Assess the success rate, number of tasks sold, and user ratings for both segments.
4. **Customer Retention Metrics**: Examine the rate of getting loyal customers and total earnings from them.
5. **Marketing Effectiveness**: Analyze the impact of promotional words, title length, and famous words on task sales.
6. **Feedback and Reviews**: Look at the number and rate of comments/reviews to gauge customer satisfaction.

By focusing on these columns and metrics, you can effectively compare the benefits of fostering customer loyalty against the efforts and costs of attracting new customers.

```python
import pandas as pd

# Define the paths to the Excel files
paths = {
    'PromotionalWordCount': 'PromotionalWordCount.xlsx',
    'aboutTitle': 'aboutTitle(EffectOfTittle_).xlsx',
    'post': 'post.xlsx',
    'aboutUser': 'aboutUser(groupbySAMEName+Cat+subCat).xlsx',
    'top_100_famous_words': 'top_100_famous_words.xlsx',
    'category': 'catagory.xlsx'
}

# Load the Excel files into DataFrames and extract the necessary columns
about_title_df = pd.read_excel(paths['aboutTitle'])[['Post_ID', 'NumFamousWord', 'PromotionalWordCount', 'lengthTitle', 'num_sold', 'repeat hired', 'rating_comment', 'Num_comment', 'cost']]
post_df = pd.read_excel(paths['post'])[['Post_ID', 'category_id', 'sub_category', 'title', 'namefreelance', 'rating_comment', 'Num_comment', 'percentsuccess', 'cost', 'num_sold', 'repeat hired']]
about_user_df = pd.read_excel(paths['aboutUser'])[['namefreelance', 'category_id', 'sub_category', 'cost_per_task', 'Total_Sum_User_Sold', 'Total_Hire_Repeat_User', 'Total_user_gain', 'RateGetting_loyaltCustomer', 'Total_user_gain_from_loyaltyCustomer', 'User_Rating_Comment', 'User_Num_Comment', 'RateOfGetingReview ']]
top_100_famous_words_df = pd.read_excel(paths['top_100_famous_words'])[['Word', 'Count']]
category_df = pd.read_excel(paths['category'])[['category_id', 'Total_Num_Sold', 'Total_Hire_Repeat', 'AVG_Num_Sold', 'AVG_Hire_Repeat', 'AVG_Hire_Repeat_Percentage', 'Total_ZeroNum_SoldAndRepeatedHire', 'Total_ZeroNum_SoldAndRepeatedHire_Percentage ', 'Max_Num_Comment', 'AVG_Num_Comment']]

# Combine the extracted columns into a new DataFrame
combined_df = pd.concat([
    about_title_df,
    post_df,
    about_user_df,
    top_100_famous_words_df,
    category_df
], axis=1)

# Ensure the 'edited' directory exists
import os
if not os.path.exists('edited'):
    os.makedirs('edited')

# Save the combined DataFrame to a new Excel file
combined_df.to_excel('edited/loyalVSnewCustomer.xlsx', index=False)

```
