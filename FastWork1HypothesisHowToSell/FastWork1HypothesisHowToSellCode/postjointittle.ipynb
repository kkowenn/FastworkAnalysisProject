```python
import pandas as pd

# Load the two datasets
posts_df = pd.read_excel('post.xlsx')
about_title_df = pd.read_excel('aboutTitle(EffectOfTittle_).xlsx')

merged_df = pd.merge(posts_df, about_title_df, on='title', how='outer', suffixes=('_post', '_title'))

# Create the 'venue' column
merged_df['venue'] = merged_df['num_sold_post'].fillna(0) * merged_df['cost_post'].fillna(0)

# Save the merged dataframe to a new Excel file (Fulljoin)
merged_df.to_excel('edited/Merged_Posts_and_Titles_with_Venue.xlsx', index=False)

merged_df.head()
```

```python
df = pd.read_excel('edited/FULLJOIN.xlsx')

# Create the new column 'CanSold'
df['CanSold'] = df['num_sold'].apply(lambda x: False if x == 0 else True)

# Group data by 'CanSold'
grouped_df = df.groupby('CanSold').size().reset_index(name='counts')

grouped_df
```

```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np

df = pd.read_excel['/Users/kritsadakruapat/Desktop/mainTable/edited/JoinAndDropDup.xlsx']
# Define the features and the target variable
features = ['NumFamousWord', 'PromotionalWordCount', 'lengthTitle', 'cost_post']
X = df[features].replace(0, np.nan).fillna(0)
y = df['CanSold']

# Initialize and fit the decision tree classifier
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X, y)

# Get feature importances
feature_importances = tree_model.feature_importances_

# Create a DataFrame for feature importances
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances}).sort_values(by='Importance', ascending=False)

importance_df
```

```python
import matplotlib.pyplot as plt
# Create a DataFrame for feature importances
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances}).sort_values(by='Importance', ascending=False)

# Plot the feature importances
plt.figure(figsize=(10, 6))
plt.bar(importance_df['Feature'], importance_df['Importance'], color='skyblue')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Feature Importances from Decision Tree Classifier')
plt.xticks(rotation=45)
plt.show()
```
