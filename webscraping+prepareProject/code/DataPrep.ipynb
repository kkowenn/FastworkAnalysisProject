```python
# Create a DataFrame
df = post

# Function to extract the desired part of the URL
def extract_part(url):
    start = len("https://fastwork.co/")
    end = url.find("?source=category_architect")
    word = url[start:end]
    return word.split('?')[0]

# Apply the function to the 'F' column and create a new 'G' column
df['sub_category'] = df['web-scraper-start-url'].apply(extract_part)

# Display the DataFrame
df.head()
```

```python
columns_to_keep = ['Post_ID','category_id',	'sub_category','title',	'namefreelance','web-scraper-order'	,'web-scraper-start-url','link', 'rating_comment','Num_comment','percentsuccess','cost','num_sold','repeat hired']
post = post[columns_to_keep]
post.head()
```

```python
output_file = "posts.xlsx"
post.to_excel(output_file, index=False, engine='openpyxl')
print(f"Merged data has been saved to {output_file}")
```

# user

```python

columns_to_keep = ['namefreelance', 'category_id','sub_category','num_sold', 'repeat hired','rating_comment', 'Num_comment','cost']
aboutUsers = post[columns_to_keep]
```

```python
aboutUsers.head()
```

```python
# Create DataFrame
df = aboutUsers

# Ensure numeric columns are of appropriate numeric type
df['num_sold'] = pd.to_numeric(df['num_sold'])
df['repeat hired'] = pd.to_numeric(df['repeat hired'])
df['rating_comment'] = pd.to_numeric(df['rating_comment'])
df['Num_comment'] = pd.to_numeric(df['Num_comment'])
df['cost'] = pd.to_numeric(df['cost'])

grouped_df = df.groupby(['namefreelance', 'category_id', 'sub_category']).agg(
    Total_Cost_userGain=pd.NamedAgg(column='cost', aggfunc='sum'),
    Total_Sum_User_Sold=pd.NamedAgg(column='num_sold', aggfunc='sum'),
    Total_Hire_Repeat_User=pd.NamedAgg(column='repeat hired', aggfunc='sum'),
    User_Rating_Comment=pd.NamedAgg(column='rating_comment', aggfunc='mean'),
    User_Num_Comment=pd.NamedAgg(column='Num_comment', aggfunc='sum'),

).reset_index()

# Calculate additional columns
grouped_df
```

```python
output_file = "aboutUser.xlsx"
grouped_df.to_excel(output_file, index=False, engine='openpyxl')
print(f"Merged data has been saved to {output_file}")
```

```python
from pythainlp import word_tokenize
from collections import Counter

# Function to split English and Thai words
def split_words(sentence):
    english_words = []
    thai_words = []
    for word in sentence.split():
        if any(char.isalpha() for char in word) and all(char.isascii() for char in word):
            english_words.extend(word.split())
        else:
            thai_words.extend(word_tokenize(word))
    return english_words + thai_words

# Split words for each title in the aboutTitle['title']
all_words = []
for title in aboutTitle['title']:
    all_words.extend(split_words(title))

# Count the frequency of each word
word_counts = Counter(all_words)

# Sort the words by frequency and get the top 100 words with their counts
top_100_words = word_counts.most_common(100)

# Create a DataFrame from the top 100 words
df_top_100 = pd.DataFrame(top_100_words, columns=['Word', 'Count'])

# Save the DataFrame to an Excel file
output_file = "top_100_famous_words.xlsx"
df_top_100.to_excel(output_file, index=False, engine='openpyxl')

print(f"Top 100 famous words have been saved to {output_file}")
```

```python
topWord = pd.read_excel("top_100_famous_words.xlsx")
```

```python
topWord["Word"]
```

```python
# Add a new column "NumFamousWord" to aboutTitle
def count_famous_words(title, top_words):
    split_title_words = split_words(title)
    return sum(word in top_words for word in split_title_words)

# Convert top words to a set for faster lookup
top_words_set = set(topWord['Word'])

# Apply the count function to each title
aboutTitle['NumFamousWord'] = aboutTitle['title'].apply(lambda title: count_famous_words(title, top_words_set))

aboutTitle.head()
```

```python
aboutTitle['lengthTitle'] = aboutTitle['title'].apply(len)
```
