```python
from pythainlp import word_tokenize
from collections import Counter
import pandas as pd

# Load the data
df = pd.read_excel("aboutTitle.xlsx")

# Define the list of words to look for
words_to_count = [
    'ลด', 'แลก', 'แจก', 'แถม', 'ถูก', 'ฟรี', 'โปรโมชั่น', 'พิเศษ', 'เท่านั้น', 'only',
    'คุ้ม', 'ด่วน', 'Save', 'exchange', 'give out', 'free', 'service fees', 'Special',
    'special', 'only', 'worth it', 'quickly', 'ไว', 'quick', 'fast'
]

# Function to split English and Thai words
def split_words(sentence):
    english_words = []
    thai_words = []
    for word in sentence.split():
        if any(char.isalpha() for char in word) and all(char.isascii() for char in word):
            english_words.extend(word.split())
        else:
            thai_words.extend(word_tokenize(word))
    return english_words + thai_words

# Initialize a list to store the counts
count_list = []

# Count occurrences of each word in the titles
for index, title in df.iterrows():
    words = split_words(title['title'])
    word_counts = Counter(words)
    row_counts = {word: word_counts[word] for word in words_to_count}
    count_list.append(row_counts)

# Create a DataFrame from the list of counts
count_df = pd.DataFrame(count_list)

# Combine the count DataFrame with the original titles
result_df = pd.concat([df['title'], count_df], axis=1)

# Save the result to an Excel file
output_file = "word_counts_in_titles.xlsx"
result_df.to_excel(output_file, index=False, engine='openpyxl')

print(f"Word counts in titles have been saved to {output_file}")


print(f"Word counts in titles have been saved to {output_file}")
```

```python
from pythainlp import word_tokenize
from collections import Counter
import pandas as pd

# Define the list of words to look for
words_to_count = [
    'ลด', 'แลก', 'แจก', 'แถม', 'ถูก', 'ฟรี', 'โปรโมชั่น', 'พิเศษ', 'เท่านั้น', 'only',
    'คุ้ม', 'ด่วน', 'Save', 'exchange', 'give out', 'free', 'service fees', 'Special',
    'special', 'only', 'worth it', 'quickly', 'ไว', 'quick', 'fast'
]

# Function to split English and Thai words
def split_words(sentence):
    english_words = []
    thai_words = []
    for word in sentence.split():
        if any(char.isalpha() for char in word) and all(char.isascii() for char in word):
            english_words.extend(word.split())
        else:
            thai_words.extend(word_tokenize(word))
    return english_words + thai_words

# Function to count occurrences of words to count in a sentence
def count_sell_words(sentence):
    words = split_words(sentence)
    word_counts = Counter(words)
    return sum(word_counts[word] for word in words_to_count if word in word_counts)

# Apply the function to each title and create a new column 'SellWord'
df['SellWord'] = df['title'].apply(count_sell_words)

# Save the updated DataFrame to the same Excel file
output_file = "myPrivateData.xlsx"
df.to_excel(output_file, index=False, engine='openpyxl')

print(f"The updated DataFrame has been saved to {output_file}")
```
